```python
from fid import calculate_fid
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# KV sim as more layers merged
models = [
    "KV_Merge_1Percent_Mean_OnKey",
    "KV_Merge_5Percent_Mean_OnKey",
    "KV_Merge_10Percent_Mean_OnKey",
    "KV_Merge_20Percent_Mean_OnKey",
    "KV_Merge_25Percent_Mean_OnKey",
    "KV_Merge_50Percent_Mean_OnKey",
]
graph_names = [
    "1",
    "5",
    "10",
    "20",
    "25",
    "50",
]

# Calculate FID
fid_values_KVMerge_per = []
for model in models:
    fid = calculate_fid(f"outputs_simple/{model}", "outputs_simple/BaseModel", method="cosine")
    fid_values_KVMerge_per.append(fid)
    print(f"Cosine sim ({model}): {fid}")

# Plot
plt.plot(graph_names, fid_values_KVMerge_per)
plt.xlabel("Merge %")
plt.ylabel("Cosine Sim")
plt.title("KV Merge - Cosine Sim vs Merge %")
plt.show()
```
